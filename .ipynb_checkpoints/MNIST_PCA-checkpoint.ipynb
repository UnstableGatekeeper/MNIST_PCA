# Case Study 2 - MNIST

```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv('mnist_train.csv')
```

```python
train_data
```

### Checking for class balance

```python
train_label_counts = train_data['label'].value_counts()
```

```python
print('Class Distribution for training data:\n ', train_label_counts)
```

```python
print('Class Balance for training data:\n', train_label_counts / train_label_counts.sum()*100)
```

### The class balance is reletively normal and there is no need to perform any stratification.
---
### Same with the testing data

```python
test_data = pd.read_csv('mnist_test.csv')
```

```python
test_data
```

```python
test_label_counts = test_data['label'].value_counts()
```

```python
print('Class Distribution for testing data:\n', test_label_counts)
```

```python
print('Class Balance for training data:\n', test_label_counts / test_label_counts.sum()*100)
```

### Similar details as training data

```python
training_X = train_data.drop('label', axis='columns')
```

```python
X_labels = training_X.columns
```

```python
training_y = train_data['label']
```

```python
testing_X = test_data.drop('label', axis='columns')
```

```python
testing_y = test_data['label']
```

### Let's scale the data using MinMax Normalization

```python
from sklearn.preprocessing import MinMaxScaler
mima_scaler = MinMaxScaler()
```

```python
mima_scaled_training_X = mima_scaler.fit_transform(training_X)
```

```python
mima_scaled_training_X = pd.DataFrame(mima_scaled_training_X, columns=X_labels)
```

```python
mima_scaled_testing_X = mima_scaler.fit_transform(testing_X)
```

```python
mima_scaled_testing_X = pd.DataFrame(mima_scaled_testing_X, columns=X_labels)
```

### Making a model using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
base_knn_model = KNeighborsClassifier(n_neighbors= 10)
```

```python
base_knn_model.fit(mima_scaled_training_X, training_y)
```

```python
training_pred_y = base_knn_model.predict(training_X)
```

```python
testing_pred_y = base_knn_model.predict(testing_X)
```

```python
from sklearn.metrics import accuracy_score
```

```python
training_accuracy = accuracy_score(training_y, training_pred_y)
```

```python
testing_accuracy = accuracy_score(testing_y, testing_pred_y)
```

```python
print(f'Accuracy on training data is: {training_accuracy}.')
```

```python
print(f'Accuracy on testing data is: {testing_accuracy}.')
```

### PCA - Principal Component Analysis

```python
from sklearn.decomposition import PCA
pca = PCA(n_components= 2).fit(mima_scaled_training_X)
```

```python
print('Components:\n', pca.components_)
print('Explained variance ratio:\n', pca.explained_variance_ratio_)

cummulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)
print('Cummulative Explained Variance:\n', cummulative_explained_variance)

X_pca = pca.transform(mima_scaled_training_X)
print('Transformed data shape: ', X_pca.shape)
```

### Finally, let's plot the PCA

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = training_y)
plt.xlabel('PC1'); plt.xticks([])
plt.ylabel('PC2'); plt.yticks([])
plt.title('2 components, captures {}% of total variation'.format(cummulative_explained_variance[1].round(4)*100))
plt.show()
```
